1.	Guess a number is a game that prompts a player to guess a number between 0 and 9, which is randomly generated by the system. When the input given by the user matches the number generated by the system then the user wins. The game should go as follows:
	
	Guess the number: 5
	Sorry, try again
	Guess the number: 3
	Sorry, try again
	Guess the number: 8
	You got it right! Congo!


```python
import random

def guess_number():
    """Guess a number between 0 and 9"""
    number = random.randint(0, 9)
    guess = None
    while guess != number:
        guess = int(input("Guess the number: "))
        if guess < number:
            print("Sorry, try again.")
        elif guess > number:
            print("Sorry, try again.")
        else:
            print("You got it right! Congratulations!")
    
guess_number()
```

2.	Make an improvement to the Guess a number game. Guide the user where they are standing and limit the number of attempts to     For example, the game should go like this:
	
Guess the number: 4
Too low
Guess the number: 9
Too high
Guess the number: 5
Sorry you Lost!


```python
import random

def guess_number_game():
    secret_number = random.randint(1, 10)
    max_attempts = 3
    attempts = 0

    print("Welcome to the Guess a Number game!")
    print("You have 3 attempts to guess the number between 1 and 10.")

    while attempts < max_attempts:
        guess = int(input("Attempt {}: Guess the number: ".format(attempts + 1)))

        if guess < secret_number:
            print("Too low")
        elif guess > secret_number:
            print("Too high")
        else:
            print("Congratulations! You guessed the number correctly.")
            return

        attempts += 1

    print("Sorry, you lost!")

guess_number_game()
```

3.	Let us make the above game a little more interesting by converting it into a gambling problem. Suppose that a player starts with Rs. 1,000. If a player can guess the number in his first chance, then he will be given a prize of Rs. 5,000, if he requires 2 attempts then he will get a prize of Rs.1,000. If he loses then he will lose Rs. 500. For example the game should go like this:

You have a cash of Rs. 1,000 with you... 
Guess the number: 8
Too high
Guess the number: 3
You have just won Rs. 1,000
Your balance: Rs. 2000

```python
import random

def guess_number_game():
    secret_number = random.randint(1, 10)
    max_attempts = 3
    attempts = 0
    balance = 1000

    print("Welcome to the Guess a Number game!")
    print("You have a cash of Rs. 1,000 with you...")

    while attempts < max_attempts:
        guess = int(input("Guess the number: "))

        if guess < secret_number:
            print("Too low")
        elif guess > secret_number:
            print("Too high")
        else:
            if attempts == 0:
                prize = 5000
            elif attempts == 1:
                prize = 1000
            else:
                prize = 0

            balance += prize
            print("Congratulations! You have just won Rs. {}. Your balance: Rs. {}".format(prize, balance))
            return

        attempts += 1

    balance -= 500
    print("Sorry, you lost Rs. 500. Your balance: Rs. {}".format(balance))

guess_number_game()
```

4.	Suppose that a player wants to play a game which requires him Rs. 1,000 to start. If the current balance in his account is less than Rs. 1,000 he needs to withdraw the extra money from his e-wallet.

Note that if the sum of money in his courrent account and the amount withdrawn is greater than or equal to Rs. 1,000 then he can start playing the game. However if the sum is less than Rs. 1,000 then the program should keep displaying the user the message “You still do not have enough money to start playing.” and keep prompting the user to withdraw money unless it crosses Rs. 1,000. Once ready, i.e. if his current account balance crosses Rs. 1,000, it will display a message “Now, you are ready to play the game.” Your program should also display the account balance and the current amount in the e-wallet.

(consider: initial account balance is Rs. 200 and money in the e-wallet is Rs. 5,000)

(Do further improvement by checking if the e-wallet balance becomes NIL, etc.)

```python
def start_game():
    account_balance = 200
    e_wallet_balance = 5000

    print("Your current account balance: Rs. {}".format(account_balance))
    print("Your current e-wallet balance: Rs. {}".format(e_wallet_balance))

    required_amount = 1000 - account_balance

    while required_amount > 0:
        if e_wallet_balance <= 0:
            print("Sorry, your e-wallet balance is insufficient to withdraw.")
            return

        print("You still do not have enough money to start playing.")
        withdraw_amount = min(required_amount, e_wallet_balance)
        e_wallet_balance -= withdraw_amount
        account_balance += withdraw_amount
        required_amount = 1000 - account_balance

        print("Withdrawn Rs. {} from e-wallet.".format(withdraw_amount))
        print("Your current account balance: Rs. {}".format(account_balance))
        print("Your current e-wallet balance: Rs. {}".format(e_wallet_balance))

    print("Now, you are ready to play the game.")
    print("Your current account balance: Rs. {}".format(account_balance))
    print("Your current e-wallet balance: Rs. {}".format(e_wallet_balance))

start_game()
```
